```python
#slicing- index,condition (values)
#location,
#index of location

#condition -. loc/iloc
df.iloc[:,df.columns=='HP'] # all row @ only HP column (with iloc row cannot be condition )
```

```python
df.iloc[:,df.columns!='HP'] # all row @ without HP column 
```

```python
df.loc[df['HP']>70,df.columns=='Speed'] #iloc(row without condition) -loc(row with condition)
```

```python
#logical operator -and (&) , or (|)
cond1=df.columns=='Speed'
cond2=df.columns=='HP'
df.loc[df['HP']>70,(cond1|cond2)]
```

```python
df['Speed']>100

df.loc[df['HP']>70,(df.columns=='Speed'|df.columns=='HP')]
```
